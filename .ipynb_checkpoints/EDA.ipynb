The below chart looking at life satisfaction is one negativie data point on the idea of coping being a driver factor in usage

```python
plt.figure(figsize=(10,6))
sns.countplot(data = e_cigarette_data, x = 'Life_Satisfaction', hue = 'E_Cig_User')

plt.title("Life Satisfaction")
plt.xlabel("E-Cigarette Usage Across Different Age Brackets")
plt.ylabel("Count")
plt.legend(title = "User and Non-Users of E-Cigarettes", labels = ["No", "Yes"])

plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
numeric_data = e_cigarette_data.select_dtypes(include='number')
```

```python
numeric_data.head()
```

```python
corr_matrix = numeric_data.corr().abs()
```

Below is a correlation heatmap

```python
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='coolwarm', annot_kws={"size": 5})
plt.show()
```

Below I was listing out any variables where the correlation was stronger than 0.5, as they are ones we would be concerned about

```python
corr_matrix_a = numeric_data.corr().abs() 

threshold = 0.4

upper = corr_matrix.where(np.triu(np.ones(corr_matrix_a.shape), k=1).astype(bool))

strong_pairs = [(col1, col2, upper.loc[col1, col2])
                for col1 in upper.columns
                for col2 in upper.index
                if pd.notnull(upper.loc[col1, col2]) and upper.loc[col1, col2] >= threshold]

for pair in strong_pairs:
    print(pair)
```
